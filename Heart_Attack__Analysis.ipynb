```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/heart.csv')
data
```

```python
df_no = data['output'].value_counts().values[1]
df_yes = data['output'].value_counts(). values[0]
no_percent = int(round(df_no/ (df_no + df_yes)*100, 0))
yes_percent = int(round(df_yes/ (df_no + df_yes)*100, 0))
print(no_percent)
print(yes_percent)
```

```python
values = [no_percent, yes_percent]
colors = ['r', 'g']
explode = [0.0, 0.02]
labels = ['No Heart attack', 'Heart attack']
plt.figure(figsize= (7,7))
plt.pie(values, colors = colors, labels= labels, explode = explode)
plt.title('Heart attack Percentage')
plt.show()
```

```python
df_male = data[data.sex == 0]
df_no = df_male['output'].value_counts().values[1]
df_yes = df_male['output'].value_counts(). values[0]
no_percent1= int(round(df_no/ (df_no + df_yes)*100, 0))
yes_percent1 = int(round(df_yes/ (df_no + df_yes)*100, 0))
print(no_percent1)
print(yes_percent1)
```

```python
df_female = data[data.sex == 1]
df_no = df_female['output'].value_counts().values[1]
df_yes = df_female['output'].value_counts(). values[0]
no_percent2 = int(round(df_no/ (df_no + df_yes)*100, 0))
yes_percent2 = int(round(df_yes/ (df_no + df_yes)*100, 0))
print(no_percent2)
print(yes_percent2)
```

```python
values = [no_percent1, yes_percent1]
colors = ['r', 'b']
explode = [0.0, 0.02]
labels = ['No Heart attack', 'Heart attack']
plt.figure(figsize= (7,7))
plt.pie(values, colors = colors, labels= labels, explode = explode)
plt.title('Heart Attack Percentage in male')
plt.show()

values = [no_percent2, yes_percent2]
colors = ['m', 'g']
explode = [0.0, 0.02]
labels = ['No Heart attack', 'Heart attack']
plt.figure(figsize= (7,7))
plt.pie(values, colors = colors, labels= labels, explode = explode)
plt.title('Heart Attack Percentage in Female')
plt.show()
```

```python
sns.distplot(data[data['output'] == 0]['chol'], label = 'No Heart Attack')
sns.distplot(data[data['output'] == 1]['chol'], label = 'Heart Attack')
plt.title('Heart Attack Distribution over Cholestrol')
plt.legend()
plt.grid()
```

```python
sns.distplot(data[data['output'] == 0]['trtbps'], label = 'No Heart Attack')
sns.distplot(data[data['output'] == 1]['trtbps'], label = 'Heart Attack')
plt.title('Heart Attack Distribution over blood pressure')
plt.legend()
plt.grid()
```

```python
pd.crosstab(data.cp, data.output).plot(kind = 'bar', figsize = (20,6))
plt.title('Heart Disease Frequency according to chest pain type')
plt.xlabel('Chest Pain Type')
plt.xticks(rotation = 0)
plt.ylabel('Frequency of Disease or not')
plt.show()
```
